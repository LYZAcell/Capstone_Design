```python
# 첫 번째 CSV 파일 읽기
df1 = pd.read_csv("noti_haksa.csv")

# 두 번째 CSV 파일 읽기
df2 = pd.read_csv("noti_haksa_2.csv")
```

```python
# 두 데이터프레임 합치기
merged_df = pd.concat([df1, df2], ignore_index=True)

# 합쳐진 데이터프레임을 CSV 파일로 저장
merged_df.to_csv("merged.csv", index=False)

merged_df
```

```python
#이미지 데이터까지 합치기
df3 = pd.read_csv("merged.csv")
df4 = pd.read_csv("haksa_img_text.csv")
```

```python
merged_df_1 = pd.concat([df3, df4], ignore_index=True)

# 합쳐진 데이터프레임을 CSV 파일로 저장
merged_df_1.to_csv("merged_all.csv", index=False)
```
