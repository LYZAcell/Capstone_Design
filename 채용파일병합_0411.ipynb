```python
df1 = pd.read_csv("noti_hiring.csv")
df2 = pd.read_csv('noti_hiring_2.csv')
nurak = pd.read_csv('nurak_hiring.csv')
```

```python
# 두 데이터프레임 합치기
merged_df = pd.concat([df1, df2], ignore_index=True)

# 합쳐진 데이터프레임을 CSV 파일로 저장
merged_df.to_csv("merged_1.csv", index=False)
```
